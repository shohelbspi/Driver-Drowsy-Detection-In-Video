```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
raw_data_dir = 'datasets'
```

```python
for datapath,dirname,filenames in os.walk(raw_data_dir):
    
    for file in tqdm([f for f in filenames if f.endswith('.png')]):
            if file.split('_')[4] == '0':
                dst_path = 'prepared_dataset/closed_eye'
                if not os.path.exists(dst_path):
                    os.makedirs(dst_path)
                shutil.copy(src=datapath + '/' + file, dst= dst_path)
                    
            elif file.split('_')[4] == '1':
                dst_path = 'prepared_dataset/opened_eye'
                
                if not os.path.exists(dst_path):
                    os.makedirs(dst_path)
                shutil.copy(src=datapath + '/' + file, dst= dst_path)
               
```
